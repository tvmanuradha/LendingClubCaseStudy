# Data Cleaning

```python
# Dropping columns containing all null values in it
df = df.dropna(axis=1, how="all")
df.shape
```

```python
# checking count of null values in each column
df.isnull().sum()
```

```python
# checking percentage of null values in each column
df[df.columns[df.isnull().any()]].isnull().sum()*100/df.shape[0]
```

```python
# Dropping columns having percentage of null values greater than 20%
df1 = ["desc","next_pymnt_d","mths_since_last_record","mths_since_last_delinq"]
```

```python
df.drop(df1,axis=1,inplace=True)
```

```python
# Dropping columns which is not important for analysis purpose
df2 = ["last_pymnt_d","last_credit_pull_d","collections_12_mths_ex_med","pub_rec_bankruptcies","chargeoff_within_12_mths","tax_liens"]
```

```python
df.drop(df2,axis=1,inplace=True)
```

```python
df.info()
```

```python
# Replacing NaN with zero in emplyee length column as it has important information
df.emp_length.fillna(int(0), inplace=True)
```

```python
# checking different values count of employee length column 
df.emp_length.value_counts()
```

```python
# Filtering  and updating employee length colum to integer as it is of Object datatype
def remove_string(x):
    x=str(x)
    s = ''.join(val for val in x if val.isdigit())
    return int(s)
df['emp_length']=df['emp_length'].apply(remove_string)
df.emp_length.dtype
```

```python
# checking different values count of employee length column after filtering
df.emp_length.value_counts()
```

```python
# checking value count of term column
df.term.value_counts()
```

```python
# Filtering  and updating term column to integer as it is of Object datatype
def remove_string(x):
    x=str(x)
    s = ''.join(val for val in x if val.isdigit())
    return int(s)
df['term']=df['term'].apply(remove_string)
df.term.dtype
```

```python
# checking value count of term column after filtering
df.term.value_counts()
```

```python
# Checking value count for interest rate column
df.int_rate.value_counts()
```

```python
# Extracting percentage sign from "int_rate" column and converting it to float value

def numeric_val(x):
    x=str(x)
    s = ''.join(val for val in x if val in "0123456789.")
    return float(s)
df['int_rate']=df['int_rate'].apply(numeric_val)
```

```python
# Filing Blank value of Revolving Utility with integer 0
df.revol_util.fillna(int(0), inplace=True)
```

```python
# checking value count of revol_util column
df.revol_util.value_counts()
```

```python
# Filtering  and updating revol_util colum to float as it is of Object datatype(by removing % sign)
def numeric_val(x):
    x=str(x)
    s = ''.join(val for val in x if val in "0123456789.")
    return float(s)
df['revol_util']=df['revol_util'].apply(numeric_val)
```

```python
# checking value count of revol_util column after filtering
df.revol_util.value_counts()
```

```python
df.info()
```

```python
# Deriven metric
```

```python
# Deriving new metrics "annual_inc_cat" after grouping the "annual_inc"
df["annual_inc_cat"]=pd.cut(df["annual_inc"],[0, 25000, 50000, 75000, 100000,120000],labels=["0-25000", "25000-50000", "50000-75000","75000-100000","100000+"])
```

```python
# Deriving new metrics "dti_cat" after grouping the "dti"
df['dti_cat'] = pd.cut(df['dti'],[0,6,12,18,24],labels=['0-6','6-12','12-18','18+'])
```

```python
# Deriving new metrics "loan_amnt_cat" after grouping the "loan_amnt rate"
df['loan_amnt_cat'] = pd.cut(df['loan_amnt'],[0,8000,16000,24000,32000],labels=['0-8000','8000-16000','16000-24000','24000+'])
```

```python
# Deriving new metrics "int_rate_cat" after grouping the "interest rate"
df['int_rate_cat'] = pd.cut(df['int_rate'], [0, 5, 10, 15, 20, 25], labels=['0-5', '5-10', '10-15', '15-20','20+'])
```

```python
# Deriving new metrics "loan_amnt_cat" after grouping the "loan_amnt rate"
df['loan_amnt_cat'] = pd.cut(df['loan_amnt'],[0,8000,16000,24000,32000],labels=['0-8000','8000-16000','16000-24000','24000+'])
```

```python
# Deriving new metrics "revol_util_cat" after grouping the "revol_util"
df['revol_util_cat'] = pd.cut(df['revol_util'],[0.0,20.0,40.0,60.0,80.0,100.0],labels=['0.0-20.0','20.0-40.0','40.0-60.0','60.0-80.0','80.0+'])
```

```python
# Changing the format of "issue_d" 
df["issue_d"]=pd.to_datetime(df.issue_d, format='%b-%y')
```

```python
# Changing the format of "issue_d" into dd/mm/yyyy format
df['issue_d'] = df['issue_d'].dt.strftime('%b/%m/%y')
df['issue_d']
```

```python
# Deriving the year information and putting it into a new column called "year"
df['year']=pd.DatetimeIndex(df['issue_d']).year
df['year']
```

```python
# Deriving the month information and putting it into a new column called "month"
df['month']=pd.DatetimeIndex(df['issue_d']).month
df['month']
```

```python
# Conerting the information of "annual_inc" to integer data type as it contains some values in float type
df["annual_inc"]=df.annual_inc.apply(lambda x : int(x))
```
